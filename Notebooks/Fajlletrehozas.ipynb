```python
lst = {'ID', 'Latitude', 'Longitude'} 
df = pd.DataFrame(columns = lst)
```

```python
for station in needed_stations.iterrows():
    for stat in needed_st:
        if station[1].ID == stat:
            df2 = {'ID': station[1].ID.astype('int'), 'Latitude': station[1].Latitude, 'Longitude': station[1].Longitude}
            df = df.append(df2, ignore_index = True)
```

```python
import sys
import os
sys.path.append(os.path.join(os.environ["SUMO_HOME"], 'tools'))
import sumolib
import pyproj

net = sumolib.net.readNet('clean_sacramento_d3_sr99_good.net.xml')
detectors = []
for station in df.iterrows():
    xy_pos = net.convertLonLat2XY(station[1].Longitude, station[1].Latitude)
    # look 10m around the position
    lanes = net.getNeighboringLanes(xy_pos[0], xy_pos[1], 10)
    # attention, result is unsorted
    bestLane = None
    ref_d = 9999.
    
    for lane, dist in lanes:
        if dist < ref_d:
            ref_d = dist
            bestLane = lane
        pos, d = bestLane.getClosestLanePosAndDist(xy_pos)
    #print(type(bestLane))
    if bestLane is None:
        print(type(bestLane))
    else:
        detectors.append(sumolib.sensors.inductive_loop.InductiveLoop(int(station[1].ID), bestLane.getID(), pos))
sumolib.files.additional.write("detectorsRaw.det.xml", detectors)
```

```python
lines = ""
with open('detectorsRaw.det.xml') as f:
    lines = f.readlines()
```

```python
linesNew = ""
count = 0

nlines = len(lines)

for l in lines:
    count = count + 1
    if "additional xmlns:xsi=\"http://www.w3.org/2001/XMLSchema-instance\" xsi:noNamespaceSchemaLocation=\"http://sumo.dlr.de/xsd/additional_file.xsd\"" in l:
        l = l.replace("additional xmlns:xsi=\"http://www.w3.org/2001/XMLSchema-instance\" xsi:noNamespaceSchemaLocation=\"http://sumo.dlr.de/xsd/additional_file.xsd\"", "a")
        linesNew = linesNew + l
        linesNew = linesNew + "\t"
    if "</additional>" in l:
        l = l.replace("</additional>", "</a>")
        linesNew = linesNew + l
    if "e1Detector" in l:
        l = l.replace("e1Detector", "detectorDefinition")
    if " freq=\"60\" file=\"NUL\" friendlyPos=\"True\"" in l:
        l = l.replace(" freq=\"60\" file=\"NUL\" friendlyPos=\"True\"", "")
        linesNew = linesNew + l
        if count + 1 != nlines:
            linesNew = linesNew + "\t"
```
